```python
import pandas as pd
import random
```

```python
prod = pd.read_csv("products_deepmatcher.csv")
unique_brands = prod['brand/name'].unique()
pair_df = pd.DataFrame(columns=['left_product','left_manufacturer','left_price','left_description','right_product','right_manufacturer','right_price','right_description'])

#Faz o pareamento de 400 produtos aleatórios que não sejam iguais
while len(pair_df) < 200:

    brand = random.choice(unique_brands)
    brand_products = prod[prod['brand/name'] == brand]

    if len(brand_products) >= 2:

        product_l, product_r = random.sample(list(brand_products.itertuples(index=False)), 2)

        pair_df = pair_df.append({
            'left_product': product_l.name,
            'left_manufacturer': brand,
            'left_price': product_l.price,
            'left_description': product_l.description,
            'right_product': product_r.name,
            'right_manufacturer': brand,
            'right_price': product_r.price,
            'right_description': product_r.description,
            'label': ''
        }, ignore_index=True)

pair_df.to_csv("matcher_sample.csv", index=False)
```
